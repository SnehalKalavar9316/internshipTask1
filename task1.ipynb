```python
import calendar
short_to_full={day[:3]:day for day in calendar.day_name}
def normalize_day_name(day):
    short=day[:3].capitalize()
    return short_to_full.get(short, day)
df['Day Name']=df['Day Name'].apply(normalize_day_name)
print(df)
```
